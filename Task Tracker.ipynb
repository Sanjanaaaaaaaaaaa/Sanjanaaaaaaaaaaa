```python
import csv
import os
from datetime import datetime

# Define the name of the CSV file to store the tasks
TASKS_FILE = "tasks.csv"

# Define the main menu function
def main_menu():
    print("**** TASK TRACKER APPLICATION ****")
    print("1. ADD A TASK")
    print("2. VIEW TASKS")
    print("3. EDIT A TASK")
    print("4. DELETE A TASK")
    print("5. EXIT")
    choice = input("ENTER THE NUMBER CORRESPONDING TO YOUR CHOICE: ")
    return choice

# Define the function to add a new task
def add_task():
    name = input("ENTER TASK NAME: ")
    description = input("ENTER TASK DESCRIPTION: ")
    due_date = input("ENTER DUE DATE (YYYY-MM-DD): ")
    with open(TASKS_FILE, "a", newline="") as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow([name, description, due_date])
    print("TASK ADDED SUCCESSFULLY!!!!")

# Define the function to view all tasks
def view_tasks():
    with open(TASKS_FILE, "r") as csvfile:
        reader = csv.reader(csvfile)
        print("{:<30} {:<50} {:<10}".format("****TASK NAME****", "****DESCRIPTION****", "**DUE DATE**"))
        for row in reader:
            name, description, due_date = row
            print("{:<30} {:<50} {:<10}".format(name, description, due_date))

# Define the function to edit an existing task
def edit_task():
    tasks = []
    with open(TASKS_FILE, "r") as csvfile:
        reader = csv.reader(csvfile)
        for row in reader:
            tasks.append(row)
    print("{:<5} {:<30} {:<50} {:<10}".format("**ID**", "****TASK NAME****", "****DESCRIPTION****", "**DUE DATE**"))
    for i, task in enumerate(tasks):
        name, description, due_date = task
        print("{:<5} {:<30} {:<50} {:<10}".format(i+1, name, description, due_date))
    choice = int(input("ENTER THE ID OF THE TASK TO EDIT:"))
    name = input("ENTER TASK NAME: ")
    description = input("ENTER TASK DESCRIPTION: ")
    due_date = input("ENTER DUE DATE (YYYY-MM-DD): ")
    tasks[choice-1] = [name, description, due_date]
    with open(TASKS_FILE, "w", newline="") as csvfile:
        writer = csv.writer(csvfile)
        for task in tasks:
            writer.writerow(task)
    print("TASK EDITED SUCCESSFULLY!!!!")

# Define the function to delete a task
def delete_task():
    tasks = []
    with open(TASKS_FILE, "r") as csvfile:
        reader = csv.reader(csvfile)
        for row in reader:
            tasks.append(row)
    print("{:<5} {:<30} {:<60} {:<10}".format("**ID**", "****TASK NAME****", "****DESCRIPTION****", "****DUE DATE****"))
    for i, task in enumerate(tasks):
        name, description, due_date = task
        print("{:<5} {:<30} {:<60} {:<10}".format(i+1, name, description, due_date))
    choice = int(input("ENTER THE ID OF THE TASK TO BE DELETED: "))
    tasks.pop(choice-1)
    with open(TASKS_FILE, "w", newline="") as csvfile:
        writer = csv.writer(csvfile)
        for task in tasks:
            writer.writerow(task)
    print("TASK DELETED SUCCESSFULLY!!!!")

# Define the main program loop
while True:
    
    choice = main_menu()
    if choice == "1":
        add_task()
        
    elif choice == "2":
        view_tasks()
        4
    elif choice == "3":
        edit_task()
        
    elif choice == "4":
        delete_task()
        
    elif choice == "5":
        print("EXITING TASK TRACKER APPLICATION...")
        break
    else:
        print("INVALID CHOICE. PLEASE TRY AGAIN.")
```
